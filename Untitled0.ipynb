```python
from sklearn.preprocessing import LabelEncoder
```

```python
# prompt: apply label encoder on above data frame

le = LabelEncoder()
for col in train_df.select_dtypes(include=['object']).columns:
  train_df[col] = le.fit_transform(train_df[col])

for col in test_df.select_dtypes(include=['object']).columns:
  test_df[col] = le.fit_transform(test_df[col])
```

```python
train_df.head()
```

```python
final_df=pd.concat([train_df,test_df],axis=0)
```

```python
final_df.shape
```

```python
final_df=final_df.loc[:,~final_df.columns.duplicated()]
```

```python
final_df.shape
```

```python
df_tain=final_df.iloc[:58645,:]
df_test=final_df.iloc[58645:,:]
```

```python
df_test.head()
```

```python
df_test.drop(['loan_status'],axis=1,inplace=True)
```

```python
df_test.head()
```

```python
X_train=df_tain.drop(['loan_status'],axis=1)
y_train=df_tain['loan_status']
```

```python
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
rf_random=RandomForestClassifier(n_estimators=100, random_state=42)
from xgboost import XGBClassifier
xgb_random=XGBClassifier(n_estimators=100, random_state=42)
```

```python
voting_classifier = VotingClassifier(estimators=[
    ('rf', rf_random),
    ('xgb', xgb_random)
], voting='soft')

voting_classifier.fit(X_train, y_train)
```

```python
y_pred = voting_classifier.predict(df_test)
```

```python
y_pred
```

```python

pred=pd.DataFrame(y_pred)
sub_df=pd.read_csv('/sample_submission.csv')
datasets=pd.concat([sub_df['id'],pred],axis=1)
datasets.columns=['id','loan_status']
datasets.to_csv('sample_submission.csv',index=False)
```
